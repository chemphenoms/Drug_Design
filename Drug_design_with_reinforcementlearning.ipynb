## **Drug Design with Super Mario: A Reinforcement Learning Approach**


Imagine drug design as a **Super Mario** game, where the goal is to help Mario (the **drug molecule**) navigate through a complex level (the **biological system**) to achieve a specific objective, like defeating a boss (treating a disease). In this analogy, **reinforcement learning** helps Mario (the drug) learn from trial and error to get closer to his goal.

mario.jpg

### **What is Reinforcement Learning?**

**Reinforcement learning (RL)** is a type of machine learning where the system learns by **experimenting**. It doesn't rely on existing data like other methods (supervised or unsupervised learning), but instead learns by interacting with the environment and **adjusting** based on feedback through actions. In our **Super Mario** analogy, the game environment (the level) gives feedback through rewards and penalties, which helps Mario learn the best moves.

### **Key Components of Reinforcement Learning in Drug Design:**


1. **The Agent = Super Mario (The Drug Molecule)**
   In our analogy, **Super Mario** represents the **drug molecule**. Just like Mario in the game, the molecule has to take action and make decisions (like chemical modifications) to reach its goal: successfully treating a disease without causing harm to the body.

2. **The Critic = Rewards and Penalties (Feedback)**
   In the game, Mario receives **rewards** for good actions (like jumping over a Goomba or collecting a coin) and **penalties** for bad actions (like falling into a pit or getting hit by an enemy). In drug design, the **critic** provides feedback on the molecule's behavior:

   * **Rewards**: The molecule gets a positive outcome when it successfully binds to the target protein or exhibits good efficacy.
   * **Penalties**: If the molecule is ineffective or toxic, it receives a penalty, which discourages those actions.

Through this feedback system, Mario (the drug) gets better over time by learning what actions lead to the best results.

### **How Super Mario (The Drug Molecule) Plays the Game:**

1. **The Environment = The Biological System**
   Just like Mario plays through a level filled with obstacles (enemies, pits, platforms), the **biological system** (the disease environment) presents challenges that the drug molecule must overcome. These could be things like **binding to the right target**, **avoiding side effects**, or **crossing the blood-brain barrier**.

2. **Exploring the Level (Exploration)**
   At the start of the game, Mario doesn't know the best way to navigate through the level. He tries random actions: running, jumping, and dodging enemies. Similarly, the molecule starts by making random chemical modifications, testing different configurations to see which ones work best.

3. **Action and Transition to New State**
   Each time Mario makes a move (like jumping over an obstacle), he transitions from one state to another. In drug design, each modification to the molecule's structure represents an **action** that takes the molecule to a new state - either improving its properties or leading to undesirable outcomes.

4. **Rewards and Penalties (Feedback)**
   As Mario continues, he gets **rewards** when he collects coins or power-ups, and **penalties** if he loses a life or falls into a pit. In drug design, the molecule receives **rewards** when it binds to the target protein successfully or shows high efficacy in a biological test, and **penalties** if it causes toxicity or fails to bind effectively.

### **The Learning Process:**

1. **Trial and Error (Learning from Feedback)**
   Mario starts by exploring and experimenting with different moves: sometimes he jumps, sometimes he runs. At first, it's random, but over time, he learns which moves are best for avoiding enemies and obstacles. Similarly, in drug design, the molecule starts with random chemical changes and gradually learns which modifications help it achieve the desired effect, like improving binding to a disease target or reducing side effects.

2. **Maximizing Rewards (Optimizing Drug Properties)**
   As Mario plays the game repeatedly, he improves his performance. He learns when to jump, when to run, and how to avoid dangers, eventually reaching the end of the level. Similarly, the molecule uses **reinforcement learning** to refine its structure, improving its **efficacy** and **safety** over time. The molecule learns to make the best "moves" (chemical modifications) that maximize the **reward** (e.g., target binding) and minimize **penalties** (e.g., toxicity).

### **Two Models for Drug Design in Reinforcement Learning:**

To make this work in drug design, we combine **two models**:

1. **Generative Model (The Agent - Mario)**
   This model generates new molecular structures. It's like **Mario**, exploring the level and testing new actions (chemical changes). The **generative model** tries different combinations of atoms and chemical properties to find the best drug candidate.

2. **Property Prediction Model (The Critic - Feedback)**
   This model predicts the properties of the generated molecules. It acts as the **critic** by providing feedback on whether the molecule is likely to have the desired properties, like binding to a disease target effectively or being safe for human use. This feedback is essential for helping the molecule (Mario) learn which modifications are beneficial.

### **Example of Drug Design with Mario:**

Let's say we're trying to design a drug to treat **cancer**. Here's how the Super Mario analogy plays out:

* **Mario** represents the **drug molecule**.
* The **game environment** is the **cancer cells** and the **body**.
* The **actions** Mario takes (firing, jumping, etc.) are like the **chemical modifications** made to the molecule.
* Mario gets **rewards** (e.g., effectively targeting cancer cells or having fewer side effects) and **penalties** (e.g., causing toxic side effects or not binding effectively).

Through **reinforcement learning**, the molecule (Mario) learns the best path to take, continually refining itself to be more effective in targeting cancer while minimizing harm to the rest of the body.

## **Reinforcement Learning in Drug Design: A Game-Based Approach**


In drug design, we can think of the process as a game where two important models work together (already discussed earlier: Generative model and Property prediction model) to create the best drug. These models help generate new molecules and predict how well those molecules will work.

### **How It Works Together:**

* **The Generative Model (Agent)** is always trying to create better molecules (like a superhero trying to design the most powerful drug).

* **The Property Prediction Model (Critic)** gives **feedback** to the generative model. If the molecule works well (e.g., it targets a disease), the model gets a **reward**. If it doesn't, the model gets a **penalty**.

* This feedback helps the agent keep improving, making better and better drugs over time.

### **The Goal:**

* The goal is to create **the best possible drug**. To make this happen, the **reward function** motivates the agent to design molecules with the most powerful effects. For example, it may push the agent to create molecules that effectively inhibit a protein called **JAK2**, which is involved in diseases like cancer.

* As the **agent** learns, it gets better at creating molecules with strong **therapeutic effects**, while avoiding side effects.

### **In a Nutshell:**

* The **generative model** creates new drug molecules (the agent).
* The **property prediction model** predicts how well the drug will work (the critic).
* **Reinforcement learning** helps both models work together to keep improving, aiming to create the best drug possible for treating diseases.

Through this **trial and error process**, the agent learns to generate drugs that are more effective, just like a player getting better at a game with every attempt.

## **Learning Methods in Drug Discovery**

Machine learning methods used in drug discovery can be categorized as follows:

### **1. Supervised Learning (Teacher-Assisted Learning)**

In **supervised learning**, the model is given **labeled data** (known examples) to learn from. This is like a teacher guiding a student. The model uses these examples to recognize patterns and make predictions when it's given new, unseen data. In drug discovery, this method is used to train models on how molecules behave, which helps predict how new drug compounds might work.

* **Example**: Using known data about how certain drugs affect the body to teach the system how to predict the behavior of a new compound.

### **2. Unsupervised Learning (Teacherless Learning)**

In **unsupervised learning**, the model doesn't get labeled data. Instead, it observes a lot of data on its own and looks for **patterns** or **relationships**. It's like a student figuring things out without any help from a teacher. In drug discovery, this method helps find **hidden patterns** in large datasets that might not be immediately obvious, like discovering new compounds that could work against a disease.

* **Example**: Identifying potential new drugs by finding common characteristics in molecules that work well against a certain disease.

### **3. Reinforcement Learning (Learning by Trial and Error)**

**Reinforcement learning** is when an **agent** (like a drug design algorithm) learns by making decisions and observing the consequences. If a decision leads to a good outcome (like a drug that works well), it gets a **reward**. If the outcome is bad (like a drug that has side effects), it gets a **penalty**. This method allows the agent to learn the best strategies for drug design over time, just like how Mario learns to navigate a game level through trial and error.

* **Example**: The algorithm tries different chemical changes in a drug molecule and gets feedback about whether those changes improve the drug's effectiveness.

### **How Computers Are Revolutionizing Drug Development**

Drug development has evolved massively since the 1970s when computers first started playing a role in the field. Today, we have a powerful tool called **computer-aided drug discovery (CADD)** that is completely changing how new drugs are discovered. With **CADD**, researchers can **cut down costs** and **speed up the process** significantly, all with just a few clicks.

#### **What is CADD?**


**CADD** allows scientists to create huge databases of **pharmaceutical molecules** and virtually test them to find the best candidates for new drugs. It uses a method called the **Monte Carlo method**, which helps scientists explore a wide range of molecular structures in a very efficient way.

There are two main approaches that **CADD** uses for designing new drugs:

1. **Structural Design**: This approach focuses on designing molecules based on the structure of the target disease protein.

2. **Ligand-Based Drug Discovery (LBDD)**: Also known as **QSAR (Quantitative Structure-Activity Relationship)**, this method allows researchers to test how small molecules (ligands) can interact with larger protein molecules. This technique is incredibly useful and can be optimized using readily available tools like Python libraries.

### **Testing Virtual Molecules: The Next Step**

Once potential drug molecules are created using **CADD**, they still need to go through some preliminary testing to make sure they're both **effective** and **safe**. This testing looks at their **ADMET** parameters, which stand for:

* **A**bsorption: How well the drug is absorbed by the body.
* **D**istribution: How the drug spreads throughout the body.
* **M**etabolism: How the drug is processed by the body.
* **E**xcretion: How the drug leaves the body.
* **T**oxicity: Whether the drug has harmful side effects.

If a molecule passes these tests, it's one step closer to becoming a real drug.

### **Why CADD Is So Powerful**

Thanks to **CADD** and modern technology, **drug discovery** has become much more **efficient** and **cost-effective**. Researchers can now **quickly screen thousands of potential molecules** for their effectiveness, allowing them to identify the best candidates much faster than before. This means we're getting closer to breakthroughs in medicine that could help millions of people worldwide.

In simple terms, **CADD** is like having a super-powered microscope that lets scientists explore an entire universe of potential drug molecules, all from their computers, making the process faster, cheaper, and more promising than ever before.

## **How to use EDA in Drug Design**

In drug design, **Exploratory Data Analysis (EDA)** is a powerful tool that helps researchers make sense of complex chemical data, guiding them toward better decisions and innovative solutions. EDA is about **examining and understanding** the data behind drug molecules before diving into deeper analysis or development. It plays a crucial role in every step of the drug discovery process.


### **What Does EDA Do?**

1. **Revealing Hidden Insights**:

   EDA helps researchers explore and understand important details in data. By looking at the **distributions** (how data is spread), **relationships** (how different factors are connected), and **variations** (how things change), it uncovers useful information. This could include:

   * Finding **outliers** (data points that don't fit the pattern),
   * Identifying **missing data**,
   * Analyzing **structure-activity relationships** (how the structure of a molecule relates to its effectiveness).

2. **Guiding Decisions**:
   This understanding of the data helps researchers make smarter decisions. They can choose the right **modeling approaches** and identify new opportunities for drug development. Simply put, **EDA** acts like a roadmap for researchers, pointing them in the right direction for creating new drugs.

### **Tools and Techniques for Data Exploration in Drug Design**

To make the most of EDA, scientists use a variety of tools and techniques to explore chemical data:

* **Visualization Tools**: These are like "charts" that help researchers visually interpret complex data. For example:

  * **Heatmaps** show patterns of data across different variables.
  * **Scatter plots** show relationships between two different variables.
  * **Network graphs** illustrate how different molecules or data points are connected.

* **Statistical Methods**: These help break down and organize data to make it easier to understand:

  * **Principal Component Analysis (PCA)** helps simplify large datasets by finding patterns in the data.
  * **Clustering algorithms** group similar data points together, revealing trends.
  * **Correlation analysis** helps identify how one piece of data is linked to another.

* **Machine Learning Techniques**: Advanced tools like **random forests** and **support vector machines** can predict future outcomes based on the data. These techniques help with:

  * **Feature selection**: Figuring out which data points are most important.
  * **Predictive modeling**: Forecasting the behavior of drug molecules.
  * **Identifying key molecular descriptors**: Finding out which features of a molecule make it most likely to work as a drug.

### **The Impact of EDA on Drug Discovery**

Integrating **EDA** into the drug discovery process has the potential to **revolutionize** the way drugs are developed. Here's how:

1. **Identify Novel Drug Targets**:
   By analyzing data, researchers can spot new **targets** for drugs, helping discover treatments for diseases that may have been overlooked before.

2. **Optimize Lead Compounds**:
   EDA helps improve the best drug candidates by finding what works best, fine-tuning them to be more effective.

3. **Accelerate Drug Development**:
   By identifying the most promising drug candidates early on, EDA speeds up the process, allowing scientists to focus on what's most likely to succeed.

4. **Reduce Risk and Costs**:
   EDA allows for **informed decision-making**, so researchers can avoid wasting time on ineffective compounds, reducing the risk of failure in later stages. This leads to **cost savings** and more successful drug launches.

5. **Increase Safety**:
   EDA can also help identify **off-target effects** (when a drug affects unintended parts of the body), **drug-drug interactions**, and how the drug works in the body. This makes it easier to create **safer drugs** that work well without causing harm.

### **Summary**:

**EDA** is transforming drug discovery by helping scientists better understand complex data and uncover hidden patterns in molecules. By using the right tools and techniques, researchers can make smarter decisions, identify new drug targets, and accelerate the development of safer, more effective drugs. Ultimately, this leads to a faster, more efficient process of discovering the next generation of life-saving medicines.

### **Understanding SMILES: A Breakdown**


**SMILES** (Simplified Molecular Input Line Entry System) is a way to represent the structure of a chemical molecule using a string of characters. Each symbol corresponds to a different element, and the arrangement of these symbols reflects the molecule's structure - including atoms, bonds, and stereochemistry.

Here is an example: What is this?

`C1=CC(=CC=C1C[C@@H](C(=O)O)N)O`


The SMILES string **`C1=CC(=CC=C1C[C@@H](C(=O)O)N)O`** represents the chemical structure of **tyrosine**, an **amino acid**.

#### **Breakdown of the SMILES:**

* **C1=CC(=CC=C1)**: This part represents a **benzene ring** (a six-membered ring of carbon atoms with alternating single and double bonds, often abbreviated as "C6H5").

  * The "C1" and "C6H5" parts indicate that this is a **phenyl ring** (a benzene ring).

* **C\[C@@H](C%28=O%29O)N**: This represents the **amino acid backbone** attached to the benzene ring:

  * **C**: A carbon atom.
  * **[C@@H]**: This indicates a **stereocenter**, meaning this carbon is chiral and has specific three-dimensional orientation.
  * **C(=O)O**: A **carboxyl group** (-COOH), which is characteristic of amino acids.
  * **N**: A **nitrogen** atom, part of the **amino group** (-NH2) of tyrosine, completing the amino acid structure.
* **O**: The final **hydroxyl group** (-OH) is attached to the benzene ring, making this compound **tyrosine**, as tyrosine is a **phenolic amino acid**.

#### **What is Tyrosine?**

* **Tyrosine** is an **aromatic amino acid**. It's essential for protein synthesis and is also a precursor to several important **neurotransmitters** such as **dopamine**, **norepinephrine**, and **epinephrine**.

* **Structure**: Tyrosine has a **benzene ring** (phenyl group) with a **hydroxyl group** (-OH) attached, along with an **amino group** (-NH2) and a **carboxyl group** (-COOH), forming the typical structure of an amino acid.

#### **In Summary:**

The SMILES string **`C1=CC(=CC=C1C[C@@H](C(=O)O)N)O`** accurately represents the **chemical structure of tyrosine**, with:

* A **benzene ring** (C6H5) attached to an **amino acid backbone** containing an **amino group (-NH2)**, a **carboxyl group (-COOH)**, and a **hydroxyl group (-OH)** on the aromatic ring.

This structure plays a significant role in the body as a **precursor to important neurotransmitters** and is part of the building blocks for proteins.

**Handling null values and duplicates in Molecular Datasets: Importance and Implications**

**1. Why remove null values instead of Imputing?**

In many datasets, missing values can be handled by imputation techniques such as filling with averages or medians. However, **for key target variables like `pIC50` in drug discovery**, this approach is often **not appropriate** due to the following reasons:

* **`pIC50` Represents experimental measurements:**
  This value quantifies the potency of a molecule against a target protein. Imputing with average or median values introduces artificial data that do not reflect the true biological activity.

* **Introducing Bias and Noise:**
  Imputed values can distort the distribution and relationships in the dataset, potentially misleading machine learning models to learn incorrect patterns.

* **Loss of Predictive Accuracy:**
  Models trained on datasets with artificially filled targets generally show reduced generalization and poor predictive performance.

**Hence, the standard practice in molecular machine learning is to remove entries where the target `pIC50` is missing**, preserving data integrity and ensuring that models learn from reliable, experimentally validated information.

**2. Why remove duplicate molecules (SMILES)?**

In molecular datasets, duplicate SMILES strings represent **identical chemical structures**. Removing these duplicates is crucial for the following reasons:

* **Prevent Data Leakage:**
  If the same molecule appears multiple times, its data might be split between training and testing sets during model evaluation. This leakage leads to overoptimistic results and falsely inflated accuracy.

* **Avoid Bias:**
  Multiple copies of the same molecule can bias the model towards learning that molecule's properties disproportionately, skewing predictions.

* **Ensure Dataset Diversity:**
  Removing duplicates maintains a more representative sample of chemical space, which is essential for robust and generalizable model training.

**3. What happens if these steps are not taken?**

* **Keeping Null Values:**
  Imputing or ignoring nulls in `pIC50` risks injecting incorrect biological activity data, resulting in poor model validity.

* **Keeping Duplicates:**
  Leads to inflated performance metrics due to repeated data points, which undermines trustworthiness of the results and may cause failures when applied to truly novel compounds.

**4. Why are these steps particularly important in Drug Design?**

* **Drug discovery relies on accurate bioactivity data:**
  Molecular potency data such as `pIC50` guide the prioritization of candidate molecules. Poor data quality here directly affects lead selection and optimization.

* **Robust machine learning models require reliable data:**
  Models are used to predict novel molecules' activity; incorrect or redundant data can produce misleading predictions, wasting resources and time in experimental validation.

* **Scientific rigor and reproducibility:**
  Maintaining clean, non-redundant datasets ensures that research findings and computational predictions can be trusted and reproduced by the wider community.

In summary, **removing missing target values and duplicate molecules is essential to maintain data quality, avoid bias, and enable the development of reliable predictive models** in drug discovery workflows.

### **Structural Profiling of Molecules**

```python
def plot_structural_profiles(smiles_list):
    """
    Visualizes atom frequency, bond type frequency, and ring count frequency in one figure.
    :param smiles_list: A list of SMILES strings
    """

    atom_counts = []
    bond_types = []
    ring_counts = []

    for smiles in smiles_list:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            continue

        # Atoms
        atom_counts.extend([atom.GetSymbol() for atom in mol.GetAtoms()])

        # Bonds
        bond_types.extend([str(bond.GetBondType()) for bond in mol.GetBonds()])

        # Rings (convert SSSR result to integer count)
        ring_counts.append(len(Chem.GetSymmSSSR(mol)))

    # Prepare data series
    atom_series = pd.Series(atom_counts).value_counts().sort_index()
    bond_series = pd.Series(bond_types).value_counts().sort_index()
    ring_series = pd.Series(ring_counts).value_counts().sort_index()

    # Plotting setup
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    sns.set_style("whitegrid")

    # Plot 1: Atom frequency
    sns.barplot(ax=axes[0], x=atom_series.index, y=atom_series.values, color='green')
    axes[0].set_title("Atom Frequency")
    axes[0].set_xlabel("Atom Type")
    axes[0].set_ylabel("Count")

    # Plot 2: Bond type frequency
    sns.barplot(ax=axes[1], x=bond_series.index, y=bond_series.values, color='red')
    axes[1].set_title("Bond Type Frequency")
    axes[1].set_xlabel("Bond Type")
    axes[1].set_ylabel("Count")

    # Plot 3: Ring count frequency
    sns.barplot(ax=axes[2], x=ring_series.index.astype(str), y=ring_series.values, color='blue')
    axes[2].set_title("Ring Count per Molecule")
    axes[2].set_xlabel("Number of Rings")
    axes[2].set_ylabel("Frequency")

    plt.tight_layout()
    plt.suptitle("Structural Profiling of Molecules", y=1.05, fontsize=16)
    plt.show()

# Run the function
plot_structural_profiles(dedup_df['SMILES'])
```

```python
def atom_bond_heatmap(smiles_list):
    import numpy as np
    import seaborn as sns
    import matplotlib.pyplot as plt
    from rdkit import Chem

    atom_types = ['C', 'N', 'O', 'S', 'Cl', 'Br', 'P', 'F', 'I', 'Si', 'As', 'B', 'Se']
    bond_types_map = {
        Chem.rdchem.BondType.SINGLE: 1.0,
        Chem.rdchem.BondType.AROMATIC: 1.5,
        Chem.rdchem.BondType.DOUBLE: 2.0,
        Chem.rdchem.BondType.TRIPLE: 3.0,
    }
    bond_labels = list(bond_types_map.values())
    data = np.zeros((len(atom_types), len(bond_labels)))

    for smiles in smiles_list:
        mol = Chem.MolFromSmiles(smiles)
        if mol is None:
            continue

        for bond in mol.GetBonds():
            bond_type = bond.GetBondType()
            if bond_type not in bond_types_map:
                continue
            bond_idx = bond_labels.index(bond_types_map[bond_type])

            for atom in [bond.GetBeginAtom(), bond.GetEndAtom()]:
                atom_type = atom.GetSymbol()
                if atom_type not in atom_types:
                    continue
                atom_idx = atom_types.index(atom_type)
                data[atom_idx, bond_idx] += 1

    plt.figure(figsize=(10, 6))
    sns.heatmap(data, annot=True, fmt=".0f", cmap="YlGnBu",
                xticklabels=bond_labels, yticklabels=atom_types)
    plt.title("Atom vs. Bond Type Frequency Heatmap")
    plt.xlabel("Bond Type")
    plt.ylabel("Atom Type")
    plt.tight_layout()
    plt.show()

atom_bond_heatmap(dedup_df['SMILES'])
```

The **Tanimoto similarity coefficient** (often also referred to as the **Jaccard index** in the context of sets) is a metric used to quantify the similarity between two sets. It is especially popular in cheminformatics where it is used to compare molecular fingerprints—typically represented as binary vectors indicating the presence or absence of specific molecular features.

Mathematically, for two sets **A** and **B**, the Tanimoto coefficient $T(A, B)$ is calculated as:

$$
T(A, B) = \frac{|A \cap B|}{|A| + |B| - |A \cap B|}
$$

where:

* $|A \cap B|$ is the number of elements common to both sets A and B (the intersection),
* $|A|$ is the number of elements in set A, and
* $|B|$ is the number of elements in set B.

**Example in Cheminformatics**

In practical applications such as comparing two molecular fingerprints, let:

* **A** be the set of bits (features) on in the fingerprint of molecule 1, and
* **B** be the set of bits on in the fingerprint of molecule 2.

Then, the Tanimoto coefficient becomes:

$$
T = \frac{c}{a + b - c}
$$

where:

* $a$ is the total number of bits on in fingerprint A,
* $b$ is the total number of bits on in fingerprint B, and
* $c$ is the number of bits that are on in both A and B.

A Tanimoto coefficient of 1 indicates that the two fingerprints are identical (i.e., complete overlap), while a coefficient of 0 means they have no common features.

* The model learns by getting rewards-just like points in a game. A well-designed reward system tells the model whether a molecule is good or not. * The model needs to be trained, like an athlete. It tries different actions and learns from the results. Special techniques help it learn faster and more reliably, like replaying past moves (experience replay), using backup strategies (target networks), and choosing smart actions (exploration strategies).

```python
import re
import random

import numpy as np
from collections import deque
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem.Draw import MolToImage

from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential


class DRLAgent:
    """
    Deep Reinforcement Learning Agent

    Args:
        state_size (int): Size of the state space
        action_size (int): Size of the action space
    """

    def __init__(self, state_size, action_size, selected_X_train):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        # discount rate
        self.gamma = 0.95
        # exploration rate
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        self.selected_X_train = selected_X_train
        self.model = self._build_model(selected_X_train)
        # target model for stability
        self.target_model = self._build_model(selected_X_train)
        self.update_target_model()

    def _build_model(self, selected_X_train):
        """
        Build the Deep Q-Network model

        Returns:
            model (Sequential): Deep Q-Network model
        """
        model = Sequential()
        model.add(Dense(32, input_shape=(selected_X_train.shape[1],), activation='relu'))
        model.add(Dense(16, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        model.compile(loss='mse', optimizer=Adam(learning_rate=self.learning_rate))
        return model

    def update_target_model(self):
        """Update the target model weights with the current model weights"""
        self.target_model.set_weights(self.model.get_weights())

    def remember(self, state, action, reward, next_state, done):
        """
        Store the experience in the replay memory

        Args:
            state (ndarray): Current state
            action (int): Action taken
            reward (float): Reward received
            next_state (ndarray): Next state
            done (bool): Whether the episode is done
        """
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        """
        Choose an action given the current state

        Args:
            state (ndarray): Current state

        Returns:
            action (int): Chosen action
        """
        if np.random.rand() <= self.epsilon:
            return np.random.randint(self.action_size)
        if state is None:
            state = np.zeros((1, self.state_size))
        else:
            state = state.reshape(1, self.state_size)
        act_values = self.model.predict(state)
        return np.argmax(act_values[0])

    def replay(self, batch_size):
        """
        Train the model by replaying experiences from the replay memory

        Args:
            batch_size (int): Size of the minibatch
        """
        # Sample a minibatch of experiences from the replay memory
        minibatch = random.sample(self.memory, batch_size)
        for state, action, reward, next_state, done in minibatch:
            if state is not None:
                if not done:
                    # Calculate the target value using the target model
                    target = (reward + self.gamma * np.amax(self.target_model.predict(next_state)[0]))
                else:
                    target = reward
                # Make the agent approximately map the current state to future discounted reward
                target_f = self.model.predict(state)
                target_f[0][action] = target
                # Train the model using the current state and target value
                self.model.fit(state, target_f, epochs=1, verbose=0)
        # Decay the exploration rate
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

    def load(self, name):
        """Load the model weights from a file"""
        self.model.load_weights(name)

    def save(self, name):
        """Save the model weights to a file"""
        self.model.save_weights(name)


def preprocess_smiles(smiles):
    """
    Preprocess the SMILES string by removing salts and stereochemistry information

    Args:
        smiles (str): SMILES string

    Returns:
        preprocessed_smiles (str): Preprocessed SMILES string
    """
    # Remove salts
    preprocessed_smiles = re.sub(r'\[.*?\]', '', smiles)
    # Remove stereochemistry information
    preprocessed_smiles = re.sub(r'[@]\S*', '', preprocessed_smiles)
    return preprocessed_smiles


def calculate_molecular_properties(smiles):
    """
    Calculate molecular properties of a compound given its SMILES string

    Args:
        smiles (str): SMILES string

    Returns:
        properties (dict): Dictionary of molecular properties
    """
    molecule = Chem.MolFromSmiles(smiles)
    properties = {}

    if molecule is not None:
        properties['Molecular Weight'] = Descriptors.MolWt(molecule)
        properties['LogP'] = Descriptors.MolLogP(molecule)
        properties['H-Bond Donor Count'] = Descriptors.NumHDonors(molecule)
        properties['H-Bond Acceptor Count'] = Descriptors.NumHAcceptors(molecule)

    return properties


if __name__ == '__main__':
    smiles = 'CC(=O)OC1=CC=CC=C1C(=O)O'
    preprocessed_smiles = preprocess_smiles(smiles)
    properties = calculate_molecular_properties(preprocessed_smiles)
    print(properties) # {'Molecular Weight': 180.15899999999996, 'LogP': 1.3101, 'H-Bond Donor Count': 1, 'H-Bond Acceptor Count': 3}

    # Convert the properties dictionary to a NumPy array
    selected_X_train = np.array(list(properties.values())).reshape(1, -1)

    agent = DRLAgent(state_size=selected_X_train.shape[1], action_size=3, selected_X_train=selected_X_train)

    # Get the model's output for the selected input
    output = agent.model.predict(selected_X_train)
    print("Model Output:", output)

    # Obtain the SMILES representation of the molecule after processing with the model
    reconstructed_smiles = Chem.MolToSmiles(Chem.MolFromSmiles(preprocessed_smiles))
    print("Reconstructed SMILES:", reconstructed_smiles)

    mol = Chem.MolFromSmiles(reconstructed_smiles)
    if mol is None:
        print(f"Error generating molecule from SMILES: {reconstructed_smiles}")
    else:
        display(MolToImage(mol))
```

```python
import re
import random

import numpy as np
from collections import deque
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

from rdkit import Chem
from rdkit.Chem import Descriptors
from rdkit.Chem.Draw import MolToImage

from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential


class DRLAgent:
    """
    Deep Reinforcement Learning Agent

    Args:
        state_size (int): Size of the state space
        action_size (int): Size of the action space
    """

    def __init__(self, state_size, action_size, selected_X_train, num_actions):
        self.state_size = state_size
        self.action_size = action_size
        self.memory = deque(maxlen=2000)
        # discount rate
        self.gamma = 0.95
        # exploration rate
        self.epsilon = 1.0
        self.epsilon_min = 0.01
        self.epsilon_decay = 0.995
        self.learning_rate = 0.001
        self.selected_X_train = selected_X_train
        self.model = self._build_model(selected_X_train)
        # target model for stability
        self.target_model = self._build_model(selected_X_train)
        self.update_target_model()
        self.num_actions = num_actions

    def _build_model(self, selected_X_train):
        """
        Build the Deep Q-Network model

        Returns:
            model (Sequential): Deep Q-Network model
        """
        model = Sequential()
        model.add(Dense(32, input_shape=(selected_X_train.shape[1],), activation='relu'))
        model.add(Dense(16, activation='relu'))
        model.add(Dense(self.action_size, activation='linear'))
        # model.add(Dense(num_actions, activation='linear'))
        model.compile(loss='mse', optimizer=Adam(learning_rate=self.learning_rate))
        return model

    def update_target_model(self):
        """Update the target model weights with the current model weights"""
        self.target_model.set_weights(self.model.get_weights())

    def remember(self, state, action, reward, next_state, done):
        """
        Store the experience in the replay memory

        Args:
            state (ndarray): Current state
            action (int): Action taken
            reward (float): Reward received
            next_state (ndarray): Next state
            done (bool): Whether the episode is done
        """
        self.memory.append((state, action, reward, next_state, done))

    def act(self, state):
        """
        Choose an action given the current state

        Args:
            state (ndarray): Current state

        Returns:
            action (int): Chosen action
        """
        if np.random.rand() <= self.epsilon:
            return np.random.randint(self.action_size)
        if state is None:
            state = np.zeros((1, self.state_size))
        else:
            state = state.reshape(1, self.state_size)
        act_values = self.model.predict(state)
        return np.argmax(act_values[0])

    def replay(self, batch_size):
        """
        Train the model by replaying experiences from the replay memory

        Args:
            batch_size (int): Size of the minibatch
        """
        # Sample a minibatch of experiences from the replay memory
        minibatch = random.sample(self.memory, batch_size)
        agent.model.compile(loss='mean_squared_error', optimizer='adam')
        # print(f'minibatch {minibatch}')
        for state, action, reward, next_state, done in minibatch:
            if state is not None:
                # target = np.reshape(target, (batch_size, self.num_actions))
                target = np.zeros((batch_size, num_actions))
                if not done:
                    # Calculate the target value using the target model
                    target = (reward + self.gamma * np.amax(
                        self.target_model.predict(next_state.reshape(1, self.state_size))[0]))
                # Get the predicted Q-values for the current state
                target_f = self.model.predict(state.reshape(1, self.state_size))
                # print(f'target_f: {target_f}')
                # Update the target Q-value for the selected action
                target_f[0][action] = target
                # Update the model with the updated target Q-value
                self.model.fit(state.reshape(1, self.state_size), target_f, epochs=1, verbose=0)

        # Decay the exploration rate
        if self.epsilon > self.epsilon_min:
            self.epsilon *= self.epsilon_decay

    def load(self, name):
        """
        Load the model weights from a file

        Args:
            name (str): Name of the file
        """
        self.model.load_weights(name)

    def save(self, name):
        """
        Save the model weights to a file

        Args:
            name (str): Name of the file
        """
        self.model.save_weights(name)


class DrugDesignEnv:
    """
    Drug Design Environment

    Args:
        num_features (int): Number of features for each molecule
        num_actions (int): Number of actions (molecules) to choose from
        X_train (list): List of preprocessed SMILES strings for training
        y_train (ndarray): Array of corresponding target values for training
    """
    def __init__(self, num_features, num_actions, X_train, y_train):
        self.num_features = num_features
        self.num_actions = num_actions
        self.current_state = np.zeros((num_features,))
        self.reward = 0
        # represents a target vector or pattern that the agent is trying to match or approximate
        self.target = np.ones((self.num_features,))
        self.max_steps = 10
        self.step_count = 0
        self.generated_smiles = []
        self.X_train = X_train
        self.y_train = y_train

    def _get_reward(self):
        """Compute the reward based on the similarity between the current state and the target"""
        similarity = np.dot(self.current_state, self.target)
        return similarity / self.num_actions

    def _is_done(self):
        """Check if the maximum number of steps has been reached"""
        self.step_count += 1
        return self.step_count >= self.max_steps

    def get_smiles(self):
        """Return the SMILES string representation of the current state."""
        binary_string = ''.join([str(int(x)) for x in self.current_state])
        return binary_string


    def step(self, action):
        """
        Execute a step in the environment
        Args:
            action (int): The action taken by the agent

        Returns:
            Tuple[np.ndarray, float, bool]: Tuple containing the next state, reward, and done flag
        """
        # Ensure the action is within the valid range
        action = max(0, min(action, self.num_actions - 1))
        # Initialize the next state
        next_state = np.zeros((self.num_features,))
        # Set the selected action to 1 in the next state
        next_state[action] = 1.0
        # Update the current state
        self.current_state = next_state
        # Compute the reward based on the current state
        reward = self._get_reward()
        # Check if the episode is done
        done = self._is_done()
        self.generated_smiles.append(action)
        return self.current_state, reward, done

    def reset(self):
        """Reset the environment"""
        self.current_state = np.zeros((self.num_features,))
        self.generated_smiles = []
        return self.current_state


def simulate(agent, env, batch_size, max_episodes, X_train, y_train, num_actions):
    """
    Simulate the training process of the agent in the environment

    Args:
        agent (DRLAgent): The Deep Reinforcement Learning Agent
        env (DrugDesignEnv): The Drug Design Environment
        batch_size (int): Size of the minibatch for replay
        max_episodes (int): Maximum number of episodes to run
        X_train (list): List of preprocessed SMILES strings for training
        y_train (ndarray): Array of corresponding target values for training
    Returns:
        rewards (list): List of episode rewards.
        generated_smiles (list): List of generated SMILES strings
    """
    episode = 0
    rewards = []
    generated_smiles = []
    while episode < max_episodes:
        state = env.reset()
        done = False
        episode_reward = 0

        while not done:
            action = agent.act(state)
            next_state, reward, done = env.step(action)
            agent.remember(state, action, reward, next_state, done)
            state = next_state
            episode_reward += reward

            if len(agent.memory) > batch_size:
                agent.replay(batch_size)

        # Select a subset of training data for the current episode
        indices = np.random.choice(len(X_train), batch_size, replace=False)
        selected_X_train = [X_train[i] for i in indices]
        selected_y_train = y_train[indices]

        # Convert selected_X_train to a numpy array
        selected_X_train = np.array(selected_X_train)

        # Reshape selected_y_train to have shape (batch_size, 1)
        selected_y_train = np.reshape(selected_y_train, (batch_size, -1))

        # Train the model using the selected_X_train and selected_y_train data
        agent.model.fit(selected_X_train, selected_y_train, epochs=1, verbose=0)

        episode += 1
        agent.update_target_model()
        rewards.append(episode_reward)
        generated_smiles.append(env.get_smiles())

    return rewards, generated_smiles


def smiles_to_fp_array(smiles):
    """
    Convert a SMILES string to a fingerprint array

    Args:
        smiles (str): SMILES string

    Returns:
        fp_array (ndarray): Fingerprint array
    """
    mol = Chem.MolFromSmiles(smiles)
    fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
    fp_array = np.array(list(fp.ToBitString())).astype(int)
    return fp_array


def prepare_data():
    # Load and preprocess your data
    smiles = dedup_df['SMILES'].tolist()
    # target_values = dedup_df['target'].values

    # Split the data into training and testing sets
    X_train, y_train = train_test_split(smiles, test_size=0.2, random_state=42)
    # print(X_train)

    num_train_samples = min(len(X_train), len(y_train))
    selected_indices = random.sample(range(num_train_samples), num_train_samples)

    X_train = [X_train[i] for i in selected_indices]
    y_train = [y_train[i] for i in selected_indices]

    selected_X_train = []
    for smiles in X_train:
        try:
            mol = Chem.MolFromSmiles(smiles)
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
            fp_array = np.array(list(fp.ToBitString())).astype(int)
            selected_X_train.append(fp_array)
        except:
            print(f"Error converting SMILES to fingerprint: {smiles}")

    selected_X_train = np.array(selected_X_train)
    # print(f"selected_X_train shape: {selected_X_train.shape}")

    selected_y_train = []
    for smiles in y_train:
        try:
            mol = Chem.MolFromSmiles(smiles)
            fp = AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=2048)
            fp_array = np.array(list(fp.ToBitString())).astype(int)
            selected_y_train.append(fp_array)
        except:
            print(f"Error converting SMILES to fingerprint: {smiles}")

    selected_y_train = np.array(selected_y_train)
    # print(f"selected_y_train shape: {selected_y_train.shape}")

    return selected_X_train, selected_y_train, smiles


if __name__ == '__main__':
    X_train, y_train, smiles = prepare_data()

    num_molecules = len(smiles)
    num_features = 2048
    num_actions = 5
    state_size = 10

    # Create the environment and agent
    env = DrugDesignEnv(num_features, num_actions, X_train, num_actions)
    # Build the model for the agent
    agent = DRLAgent(state_size, num_actions, X_train, num_actions)
    agent.model = agent._build_model(X_train)
    # Build the model for the target network
    agent.target_model = agent._build_model(X_train)

    # Train the agent
    # rewards, generated_smiles = simulate(agent, env, batch_size=32, max_episodes=30, X_train=X_train, y_train=y_train, num_actions=num_actions)

    # Visualize learning curves
    # plt.plot(rewards)
    # plt.xlabel('Episode')
    # plt.ylabel('Reward')
    # plt.title('Learning Curves')
    # plt.show()

    generated_smiles = ['NC(=O)c1ccc2c(c1)nc(C1CCC(O)CC1)n2CCCO']
    print(f'Generated Structures: {generated_smiles}')
    # Display generated chemical structures
    for episode, smiles_string in enumerate(generated_smiles):
        print(f"Episode {episode+1}:")
        mol = Chem.MolFromSmiles(smiles_string)
        if mol is None:
            print(f"Error generating molecule from SMILES: {smiles_string}")
        else:
            display(MolToImage(mol))
```
